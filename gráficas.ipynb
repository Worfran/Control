```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion(x):
    a = np.sqrt(x**10 + 5)
    b = np.sqrt(x**12 + (12-x**2)**2)
    
    return (a/b)
```

```python
tiempos = np.linspace(0,100000,10000)
```

```python
def fase(x):
    a = np.arctan(x)
    b = x*((-x**2)/(2-3*x**2))
    c = np.arctan(b)
    
    respuesta = a-c
    return respuesta
```

```python
respuestas = []
respuestas_2 = []
for x in tiempos:
    i = funcion(x)
    respuestas.append(i)
    
    f = fase(x)
    respuestas_2.append(f)
    
```

```python
plt.figure(figsize = (5,5))
plt.title("Comportamiento de la amplitud de la corriente \n en función de la frecuencia angular (w)")
plt.xlabel("Frecuencia angular(w)")
plt.ylabel("Corriente en la inductancia (I)")
plt.plot(tiempos,respuestas)
plt.savefig("Comportamiento cualitativo", dpi = 300)
```

```python
plt.figure(figsize = (5,5))
plt.title("Comportamiento de la fase de la corriente \n en función de la frecuencia angular (w)")
plt.xlabel("Frecuencia angular(w)")
plt.ylabel("Fase de la corriente")
plt.plot(tiempos,respuestas_2)
plt.savefig("Comportamiento cualitativo", dpi = 300)
```

```python
def amplitud_numerica(x):
    rta = np.sqrt((x**2/4000000 + 1)/((10000 - (3*x**2)/100000)**2 + x**2*(1/50 - x**2/200000000)**2))
    return rta
```

```python
def fase_numerica(x):
    
    respuesta = np.arctan(x/2000) - np.arctan((x*(-x**2/200000000 - 62/25))/(10000 - (3*x**2)/100000))

    

    return respuesta
```

```python
corriente = []
desfase = []

for x in tiempos:
    i = amplitud_numerica(x)
    corriente.append(i)
    
    f = fase_numerica(x)
    desfase.append(f)
    
```

```python
w_resonancia = np.argmax(corriente)
```

```python
w = tiempos[w_resonancia]
print(w)
```

```python
corte = np.ones(len(tiempos))
resonancia = corte*w
```

```python
plt.figure(figsize = (5,5))
plt.xlabel("w(Hz)")
plt.ylabel("I(A)")
plt.plot(tiempos, corriente)
plt.legend()
plt.savefig("Comportamiento númerico", dpi = 300)
```
